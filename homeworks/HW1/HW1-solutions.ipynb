### A1

Show that:
$$
\sum_{i=1}^n (x_i - \bar{x}) = 0
$$
where $\bar{x}$ is a sample mean (arithmetic average). Show that the same holds for $\mathbb{E}[X$ and $\mathbb{E}[X]]$, that is:
$$
\mathbb{E}[X - \mathbb{E}[X]] = 0
$$

#### SOLUTION

##### Part I

$$
\begin{alignedat}{3}
    \sum_{i=1}^n (x_i - \bar{x})
    &= \sum_{i=1}^n x_i - \sum_{i=1}^n \bar{x}
    \qquad&\text{(addition is commutative)} \\
    &= n \times \frac{1}{n} \sum_{i=1}^n x_i - \sum_{i=1}^n 1 \times \bar{x} \\
    &= n\bar{x} - \bar{x}\sum_{i=1}^n 1 \\
    &= n\bar{x} - n\bar{x} \\
    &= 0
\end{alignedat}
$$

##### Part II

$$
\begin{alignedat}{3}
    \mathbb{E}[X - \mathbb{E}[X]]
    &= \mathbb{E}[X] - \mathbb{E}[\mathbb{E}[X]]
    \qquad&\text{(linearity of $\mathbb{E}$)} \\
    &= \mathbb{E}[X] - \mathbb{E}[X]
    \qquad&\text{($\mathbb{E}$ is a constant)} \\
    &= 0
\end{alignedat}
$$

### A4

Let $A$ and $B$ be two events and $P(A \cap B) = P(A)P(B)$. Does it imply that:
$$
P(A^C \cap B^C) = P(A^C)P(B^C)
$$
where $A^C$ and $B^C$ are complements of events $A$ and $B$. Justify your answer (mathematically).

#### SOLUTION

In this derivation we will use one of De Morgan's laws:
$$
A^C \cap B^C = (A \cup B)^C
$$

$$
\begin{alignedat}{3}
       P(A^C \cap B^C)
       &= P[(A \cup B)^C] \\
       &= 1 - P(A \cup B) \\
       &= 1 - [P(A) + P(B) - P(A \cap B)] \\
       &= 1 - P(A) - P(B) + P(A \cap B) \\
       &= (1 - P(A)) - P(B) + P(A)P(B)
       \qquad&\text{($A$ and $B$ are independent)} \\
       &= P(A^C) - P(B) + P(A)P(B) \\
       &= P(A^C) - P(B)(1 - P(A)) \\
       &= P(A^C) - P(B)P(A^C) \\
       &= P(A^C)(1 - P(B)) \\
       &= P(A^C)P(B^C)
\end{alignedat}
$$

Thus, $A^C$ and $B^C$ are also idependent.

### A5

Let $\bar{x}_1 = 80$ be a sample mean in a group of $n_1 = 50$ observations ($s_1^2 = 36$) and $\bar{x}_2 = 85$ be a sample mean in a group of $n_2 = 60$ observations ($s^2_2 = 49$). Use an appropriate two-sample $z$ test to test whether true means in the two groups are different. Start by defining proper null and alternative hypotheses. Provide $p$-value and an appropriate confidence interval. Use significance level $\alpha = 0.01$.

#### SOLUTION

We are interested in the following hypotheses:

$$
H_0: \mu_1 = \mu_2 \quad\Longrightarrow\quad \mu_1 - \mu_2 = 0
$$
$$
H_1: \mu_1 \neq \mu_2 \quad\Longrightarrow\quad \mu_1 - \mu_2 \neq 0
$$

Thus, since we can express the hypotheses in terms of the difference between the true means it is reasonable
to use the following estimate:
$$
\bar{d} = \bar{x}_1 - \bar{x}_2 = -5
$$

Since the groups are independent, the rules governing variances of sums of random variables tells us that
the variance of the estimate $\bar{d}$ can be estimated as:
$$
s^2_{\bar{d}} = \frac{s^2_1}{n_1} + \frac{s^2_2}{n_2}
$$

And the standard deviation of $\bar{d}$ (standard error) is:
$$
s_{\bar{d}} = \sqrt{\frac{s^2_1}{n_1} + \frac{s^2_2}{n_2}} \approx 1.24
$$

Hence, we can construct the following test statistic $z$ which will have asymptotically normal distribution (and since sample sizes are moderately large
we can assume that this approximation is quite accurate):
$$
z = \frac{\bar{d}}{\sqrt{s^2_1/n_1 + s^2_2/n_2}}
$$

So substituting the actual values we get:
$$
z = \frac{-5}{\sqrt{36/50 + 49/60}} =  \frac{-5}{\sqrt{0.72 + 0.817}} \approx \frac{-5}{1.24} \approx -4.03
$$

We significance level $\alpha = 0.01$ and a two-sided hypothesis. Thus, the rejection region should be sensitive to both very low and very high values of $z$.
It has the following form:

$$
\begin{alignedat}{3}
    R 
    &= (-\infty, q_Z^{0.005}] \cup [q_Z^{.995}, \infty) \\
    &\approx (-\infty, -2.58] \cup [2.58, \infty)
\end{alignedat}
$$
where $q_Z^{0.005}$ and $q_Z^{0.995}$ are $0.005$ and $0.995$ quantiles of the standard normal distribution.

In other words, we have that if $z \in R$ we can reject $H_0$ at the significance level $\alpha = 0.01$. Clearly, $z < 2.58$ so $z \in R$ and therefore
we can reject $H_0$ and assume $H_1$.

However, it would be nice to have a more finely-grained measure of the strength of evidence against $H_0$. To do this, we can compute $p$-value which
is equivalent to the largest level of signficance $\alpha$ at which we would still be able to reject $H_0$. We derive $p$-values directly from the distribution
of the test statistic and the observed value of $z$.

$$
p = P(Z \leq -|z|) + P(Z \geq |z|) = 2 \times P(Z \leq |z|) = 2 \times (1 - \Phi(|z|)) \approx 5.5 \times 10^{-5} < 0.001
$$

We see that the $p$-value is very low, much lower than $1 / 1000$. In such cases people often just say that $p$-value is lower than $0.001$, 
although this tend to depend on customs in a particular scientific community or a journal.

Lastly, we should also give some indication of the size of the observed effect that takes into account the uncertainty of the estimation procedure.
In other words, we should compute a confidence interval, which should be appropriate in the context of the test we run. Hence, here we should
compute a two-sided confidence interval with confidence level $1 - \alpha = 0.99$:

$$
\begin{alignedat}{3}
    \text{CI}_{\bar{d}}^{0.99} 
    &= \left[\bar{d} + q_Z^{0.005}s_{\bar{d}},\quad \bar{d} + q_Z^{0.995}\right] \\
    &\approx \left[\bar{d} - 2.58 \times s_{\bar{d}},\quad \bar{d} + 2.58 \times s_{\bar{d}}\right] \\
    &\approx \left[-5 - 3.20,\quad -5 + 3.20\right] \\
    &= \left[-8.20,\quad -1.80\right]
\end{alignedat}
$$

So we can say that with the confidence level of $0.99$ we can claim that the true difference between the means is somewhere between $-8.20$ and $-1.80$.
But remember that it does not mean that we $99\%$ probability that this happens, since the true difference is fixed and it is either in this range or not.
The confidence level corresponds to the frequency with which our interval would contain the true difference if we repeated the study multiple times.

**REPORTING**

The mean in the group I was $80$ (SD = $6$, $n = 50$) and in the group II was $85$ (SD = $7$, $n = 60$). To test whether the means are different a two-sided asymptotic $z$ test has been conducted (the sample size justified this choice). Significance level $\alpha = 0.01$ was used.
The test yielded significant results, $z = -4.03, p < 0.001$, showing that the mean in group II is significantly higher. A two-sided $99\%$ confidence interval
for the difference between the means was $[-8.20,\quad -1.80]$.

In general, non-parametric tests seem to be super nice as they use fewer assumptions, but in practical terms they

a simpler estimation in terms of averages. And the reason for that is again simple. Pseudomedian is a complicated
quantity that is diffcult to handle analytically and averages are simple, so it is easier to use them in later
stages of statistical modeling.

````r
# IMPLEMENTATION BASED ON BUILT-IN FUNCTION
pairwise.t.test(chickwts$weight, chickwts$feed, p.adjust.method = "none")
````

So above you have examples of how you can run tests like that. The second approach is of course recommended
as it is easier and less error-prone, since you do not have to write any code yourself.

However, the point of this exercise was to think about what we did here and answer whether it makes sense.

And the answer is that it does not make sense. If we assume that there are no significant effects ($H_0$ holds everywhere)
and that all tests are independent (which can not be perfectly true in this case, but let us use this as a simplifying
assumption) we have to realize that in such a cace, when there is nothing to really to discover as all means are equal,
we would have a high probability of observing at least one significant result purely by chance. In general,
when we run multiple tests like here, the type I error rate will always increase.

With our simplifying assumptions we can even calculcate the effective type I error rate.

Assuming $\alpha = 0.05$ the probability of observing a significant result even though $H_0$ really holds is $5\%$.
So if tests are independent, the probability of observing one or more significant results even though $H_0$ is true
in all cases is ($X$ denote the number of significant results):

$$
\begin{alignedat}{3}
    P(X \geq 1 | H_0\text{ is true everywhere})
    &= 1 - P(X = 0 | H_0\text{ is true everywhere}) \\
    &= 1 - \prod_{i=1}^n P(X_i = 0 | H_0\text{ is true}) \\
    &= 1 - (1 - \alpha)^n
\end{alignedat}
$$

So in our case the probability of observing at least one type I error is $1 - 0.95^{15} \approx 0.54$.
We see it is really huge, much larger than $0.05$ that we assumed. This is why in the context multiple testing
$p$-values have to be properly adjusted. We will discuss this problem later in the course.

Check whether sepals of _setosa_ flowers are on average longer in than those of _virginica_. Choose a proper test and hypothesis. Use $\alpha = 0.01$.

**NOTE.** There are good ways to visualize results. Anything that is legible and presents the results will be accepted.

#### SOLUTION

This was a somewhat trick question. First let us look at means:

````r
iris %>%
    filter(Species %in% c("setosa", "virginica")) %>%
    group_by(Species) %>%
    summarize(
        M = mean(Sepal.Length),
        SD = sd(Sepal.Length)
    )
````

Clearly, _setosa_ flowers have on average shorter sepals than _virginica_ flowers, so there is no way a test of hypothesis:

$$
H_1: \mu_{\text{setosa}} > \mu_{\text{virginica}}
$$

can yield significant results.

So there were few equally good ways to proceed:

 1. Conduct a test with the above alternative hypothesis. The result will trivially insignificant, but it is allright.
 2. Decide to run a two-sided test to check if the difference between _setosa_ and _virginica_ is significant.
 
But it was also possible to make an error. Specifically, an error would be to decide to run a one-sided test but with 
a reversed direction, that is:

$$
H_1: \mu_{\text{virginica}} > \mu_{\text{setosa}}
$$

It would be an error, because we would make this decision **after looking at the data**. So this would make our
test invalid and the true type I error rate would be larger than the one officially used. The reason is that
if we take such a decision after looking at data we effectively introduce additional information which is not
accounted for in calculations. This is exactly like the overfitting problem when training predictive models.

Below is the code for running the first (correct) one-sided hypothesis.

````r
data <- filter(iris, Species %in% c("setosa", "virginica"))

data %>%
    ggplot(aes(sample = Sepal.Length)) +
    geom_qq() +
    geom_qq_line() +
    facet_wrap(~Species) +
    labs(x = "Theoretical quantiles", ylab = "Observed quantiles")


## PERSONALLY I WOULD NOT MAKE A BIG DEAL OUT OF DEVIATIONS LIKE THAT AND JUST ASSUME NORMALITY

# ONE-SIDED TEST (setosa > virginica)
t.test(Sepal.Length ~ Species, data = data, var.equal = FALSE, alternative = "greater")
````

Clearly, the $p$-value is equal to $1$, so the results are perfectly non-significant.

In this exercise you will work with `diamonds` dataset. It consists of many different measurements for a set $53 940$ diamonds of different types and quality.
However, here you are interested in the correlation between `carat` and `price`.

````r
head(diamonds)
````

Consider _cuts_ of diamonds which is a jargon term of general quality.

````r
table(diamonds$cut)
````

First, create a new column in which you define a simplified classification based on cuts. Specifically, you should collapse `Fair`, `Good` and `Very Good` together and join `Premium` and `Ideal` so they make a second class. Use your new classification to test if the correlation between `carat` and `price` is the same in both groups.

Try to report your results in a reasonable way. They should be more or less understandable for a lay person, but also provide more finely-grained details. In particular, you should somehow show how much uncertainty there is around your best point estimate.

**HINT.** You can compute correlation between two data vectors with `cor` function.

**HINT II.** This is a sort of problem which can be hard to tackle analytically. It may be better to use a more computational approach.

````r
## FIRST COMPUTE THE NEW GROUPING VARIABLE

diamonds$superb <- ifelse(diamonds$cut %in% c("Premium", "Ideal"), "Superb", "Other")
table(diamonds$superb)
````

It is good to take a look on the joint distribution of `carat` and `price` in the groups.

````r
diamonds %>%
    ggplot(aes(x = carat, y = price)) +
    geom_point(alpha = .25) +
    facet_wrap(~superb)
````

The joint distribution seem to quite curved and rather are not jointly normal, so perhaps it is more justified
to use a rank correlation here (which will capture curvilinear monotonic relationships better).

````r
diamonds %>%
    group_by(superb) %>%
    summarize(cor = cor(carat, price, method = "spearman"))
````

Both correlations are very high and actually seem to be very close to each other. Can we judge whether the observed
difference is significant or not?

There is no simple analytic test for a difference between two Spearman correlations in two independent groups,
so the situation seems hopeless. However, this is a perfect moment to apply bootstrap with which we can compute
anything we want. We will use bootstrap to simulate the sampling distribution of the difference between Spearman
correlation coefficient in the two groups and then we will use this simulated distribution to calculate a $95\%$
percentile confidence interval and (if the sampling distribution is approximately normal) conduct a simple Wald-type test.

````r
## SET RANDOM SEED FOR REPLICABILITY
set.seed(100)
## NAIVE IMPLEMENTATION WITH A FOR-LOOP
R <- 1000  # number of bootstrap replicates

# Pre-initialized vector for storing difference between correlations
d_rho_boot0 <- vector(mode = "numeric", length = R)

for (i in 1:R) {
    boot_idx <- sample(1:nrow(diamonds), size = nrow(diamonds), replace = TRUE)
    df1 <- diamonds %>%
        slice(boot_idx) %>%
        filter(superb == "Superb")
    df2 <- diamonds %>%
        slice(boot_idx) %>%
        filter(superb == "Other")
    rho1 <- cor(df1$carat, df1$price, method = "spearman")
    rho2 <- cor(df2$carat, df2$price, method = "spearman")
    d_rho_boot0[i] <- rho1 - rho2
}
````

````r
summary(d_rho_boot0)
````

````r
set.seed(303)
## IMPLEMENTATION USING `boot` package
d_rho_boot <- boot(
    data = diamonds,
    statistic = function(data, idx) {
        df1 <- slice(data, idx) %>%
            filter(superb == "Superb")
        df2 <- slice(data, idx) %>%
            filter(superb == "Other")
        
        rho1 <- cor(df1$carat, df1$price, method = "spearman")
        rho2 <- cor(df2$carat, df2$price, method = "spearman")
        rho1 - rho2
    },
    R = 1000,
    sim = "ordinary"
)
````

````r
summary(d_rho_boot$t[,1])
````

We see that distributions in both cases are very similar.

We will procedd further using the second one.

Now, once we simulated the sampling distribution of the difference between the correlation coefficients
we can use it calculate a $95\%$ percentile confidence interval very easily.

````r
print(quantile(d_rho_boot$t[,1], probs = c(0.025, 0.975)))
````

We see that the confidence interval does not contain $0$, so we can already say that there is evidence
to claim (with $\alpha = 0.05$) that monotonic correlation between `price` and `carat` is slightly weaker
for _superb_ diamonds.

This is an overkill, since we already have an aswer, but if we need, we can also compute a $p$-value with a Wald test.
We can do that because the sampling distribution is approximately normal.

````r
qqnorm(d_rho_boot$t[, 1]); qqline(d_rho_boot$t[, 1])
````

So we can construct a standard Wald-type test statistic:

$$
\text{TS} = \frac{\rho_{\text{superb}} - \rho_{\text{other}}}{s_{\rho}^{\text{boot}}}
$$

````r
ts <- d_rho_boot$t0 / sd(d_rho_boot$t[, 1])

pval <- 2 * (1 - pnorm(abs(ts)))
pval
````

Clearly, $p$ is much lower than $0.05$, so we can reject $H_0$.

**STATISTICAL VS SCIENTIFIC SIGNIFICANCE**

Although above we could show that the difference is significant it can be debated whether a result like that
is really important. In both cases the strength of correlation is very strong and relatively to the strength of
the relationshipe the difference between groups is very weak. Moreover, the fact that we can detect this difference
and consider it _significant_ is largely an effect of a very large sample size ($n = 53940$) and not a strong
difference between the groups. Thus, it can be argued that such a result, even if signficiant, is of little scientific
importance.

It is important that one is always concsious of what one is doing and try to interpret the results in a substantial
way. It is a bad practice to try to report and everything, even when observed effects are very small, since with
large enough sample size any difference (however infinitesmal) may be shown to be _statistically significant_
as there are rarely in nature any difference of exactly $0$.

Therefore, always think about the meaning of results instead of automatically consider anything that gives you
significant results important.
